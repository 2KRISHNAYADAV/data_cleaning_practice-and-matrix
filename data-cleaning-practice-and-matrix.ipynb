# Removing Duplicates

```python
file_path = '/kaggle/input/data-cleaning-practice/data_cleaning_practice.xlsx'  
data = pd.read_excel(file_path)

# Drop duplicates
data = data.drop_duplicates()

# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns

# Display the entire DataFrame
print(data)
```

# Handling Missing Values

```python


# Fill missing values in 'Age' column with the median value
data['Age'] = data['Age'].fillna(data['Age'].median())

pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns

# Display the entire DataFrame
print(data)

```

# For Gender, fill with the mode

```python
data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns

# Display the entire DataFrame
print(data)
```

# For Salary, convert to numeric, errors='coerce' will turn invalid entries to NaN

```python
data['Salary'] = pd.to_numeric(data['Salary'], errors='coerce')
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns

# Display the entire DataFrame
print(data)
```

#  delete rows containing NaN (missing) values in your DataFrame

```python
data['Salary'] = pd.to_numeric(data['Salary'], errors='coerce')

# Drop rows with any NaN values
data = data.dropna()


pd.set_option('display.max_rows', None)  
pd.set_option('display.max_columns', None)  

# Display the entire DataFrame
print(data)
```


# For Join Date, convert to datetime

```python
data['Join Date'] = pd.to_datetime(data['Join Date'], errors='coerce')
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_columns', None)  

print(data)
```

# Drop rows with any missing values (or handle as needed)

```python
data = data.dropna()
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_columns', None)  

print(data)
```

#  Standardize the Gender colum

```python
data['Gender'] = data['Gender'].replace({'M': 'Male', 'F': 'Female', 'Unknown': 'Other'})
# pd.set_option('display.max_rows', None)  
# pd.set_option('display.max_columns', None)  

print(data)
```

# Fix invalid salary entries by replacing NaN with the mean salary

```python
data['Salary'] = data['Salary'].fillna(data['Salary'].mean())
print(data)
```

# Validate Feedback Scores

```python
data['Feedback Score'] = pd.to_numeric(data['Feedback Score'], errors='coerce')
data = data[data['Feedback Score'].between(1, 5)]  # Keep scores between 1 and 5
print(data)
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python


# Histogram of Age
fig_age = px.histogram(data, x='Age', title='Distribution of Age', nbins=10)
fig_age.show()

# Salary Distribution
fig_salary = px.histogram(data, x='Salary', title='Salary Distribution', nbins=20)
fig_salary.show()

#  Bar Chart of Gender Distribution
gender_counts = data['Gender'].value_counts().reset_index()
gender_counts.columns = ['Gender', 'Count']  # Rename the columns for clarity

fig_gender = px.bar(gender_counts, 
                    x='Gender', 
                    y='Count', 
                    title='Gender Distribution', 
                    labels={'Gender': 'Gender', 'Count': 'Count'})
fig_gender.show()

# Box Plot of Salary by Gender
fig_box = px.box(data, x='Gender', y='Salary', title='Box Plot of Salary by Gender')
fig_box.show()

# Scatter Plot of Age vs. Salary
fig_scatter = px.scatter(data, x='Age', y='Salary', color='Gender', 
                         title='Scatter Plot of Age vs. Salary')
fig_scatter.show()

# Time Series of Employees Joining by Year
data['Join Year'] = pd.to_datetime(data['Join Date']).dt.year  # Extracting the year from Join Date
fig_time_series = px.histogram(data, x='Join Year', title='Number of Employees Joined by Year', 
                                histfunc='count', 
                                labels={'Join Year': 'Year', 'count': 'Number of Employees'})
fig_time_series.show()
```
